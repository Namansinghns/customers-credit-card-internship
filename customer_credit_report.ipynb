```python
np.unique_values(customer["Customer_Age"])  # income by age group
```

```python
def age_group(age):
  if age <= 30:
    return "21-30"
  elif age <= 40:
    return "31-40"
  elif age <= 50:
    return "41-50"
  elif age <= 60:
    return "51-60"
  elif age <= 70:
    return "61-70"
  else:
    return "71+"
customer["age_group"] = customer["Customer_Age"].apply(age_group)
```

```python
customer.groupby("age_group")["Income"].mean().sort_index(ascending=True)
```

```python
credit["month"] = pd.to_datetime(credit["Week_Start_Date"]).dt.month  # income by gender and date
credit["year"] = pd.to_datetime(credit["Week_Start_Date"]).dt.year
```

```python
merge = customer.merge(credit, on='Client_Num')
```

```python
data = merge.groupby(["Gender","year","month"])["Income"].sum().unstack(0)
```

```python
plt.figure(figsize=(10, 6),facecolor="lightgreen")
data.plot(marker = "o")
plt.xlabel('Gender')
plt.ylabel('Income')
plt.title('Income by Gender')
plt.grid(True)
plt.show()
```

```python
Customer_Job = customer["Customer_Job"].values    # income by customer job
Income = customer["Income"].values

unique_job , indices = np.unique(Customer_Job, return_inverse=True)
income_sum = np.zeros(len(unique_job))

for i in range(len(Income)):
  income_sum[indices[i]] += Income[i]

for j, val in zip(unique_job, income_sum):
  print(j,":",val)
```

```python
job_sorted = np.argsort(income_sum)[::-1]
unique_job = unique_job[job_sorted]
income_sum = income_sum[job_sorted]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(unique_job, income_sum/1e6)
plt.xlabel('Customer Job')
plt.ylabel('Income')
plt.title('Income by Customer Job')
plt.show()
```
